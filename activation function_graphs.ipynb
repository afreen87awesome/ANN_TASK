```python
#sigmoid with derivatives 
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of values
x = np.linspace(-10, 10, 400)

# Compute the sigmoid function directly
sigmoid_x = 1 / (1 + np.exp(-x))

# Compute the derivative of the sigmoid function directly
sigmoid_derivative_x = sigmoid_x * (1 - sigmoid_x)

# Plot sigmoid function
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, sigmoid_x, label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('(x)')
plt.grid(True)
plt.legend()

# Plot derivative of sigmoid function
plt.subplot(1, 2, 2)
plt.plot(x, sigmoid_derivative_x, label='Sigmoid Derivative', color='orange')
plt.title('Derivative of Sigmoid Function')
plt.xlabel('x')
plt.ylabel('\'(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Tanh with derivatives 
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-5, 5, 400)

# Compute y values for tanh and its derivative directly
y_tanh = np.tanh(x)
y_tanh_derivative = 1 - np.tanh(x)**2

# Create the plots
plt.figure(figsize=(12, 6))

# Plot tanh function
plt.subplot(1, 2, 1)
plt.plot(x, y_tanh, label='tanh(x)', color='blue')
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.legend()

# Plot derivative of tanh function
plt.subplot(1, 2, 2)
plt.plot(x, y_tanh_derivative, label='Derivative of tanh(x)', color='red')
plt.title('Derivative of Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('Derivative')
plt.grid(True)
plt.legend()

# Show plots
plt.tight_layout()
plt.show()
```

```python
#relu  with derivatives 
import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-10, 10, 400)

# Compute the ReLU function and its derivative directly
y_relu = np.maximum(0, x)
y_relu_derivative = np.where(x > 0, 1, 0)

# Plot the ReLU function
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, y_relu, label='ReLU')
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.legend()

# Plot the derivative of the ReLU function
plt.subplot(1, 2, 2)
plt.plot(x, y_relu_derivative, label='ReLU Derivative', color='r')
plt.title('Derivative of ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('ReLU\'(x)')
plt.grid(True)
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

```python
#reaky_relu with derivatives 

import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-10, 10, 400)

# Compute the Leaky ReLU function and its derivative directly
alpha = 0.01
y_leaky_relu = np.where(x > 0, x, alpha * x)
y_leaky_relu_derivative = np.where(x > 0, 1, alpha)

# Plot the Leaky ReLU function
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, y_leaky_relu, label='Leaky ReLU')
plt.title('Leaky ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('Leaky ReLU(x)')
plt.grid(True)
plt.legend()

# Plot the derivative of the Leaky ReLU function
plt.subplot(1, 2, 2)
plt.plot(x, y_leaky_relu_derivative, label='Leaky ReLU Derivative', color='r')
plt.title('Derivative of Leaky ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('Leaky ReLU\'(x)')
plt.grid(True)
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

```python
#softmax with derivatives 
import numpy as np
import matplotlib.pyplot as plt

# Generate input data
z = np.linspace(-5, 5, 100)
z = np.vstack([z, np.zeros_like(z)])  # Two classes

# Compute Softmax
exp_z = np.exp(z - np.max(z))  # Subtract max for numerical stability
softmax_output = exp_z / np.sum(exp_z, axis=0)

# Compute Softmax derivative
S = np.diagflat(softmax_output) - np.outer(softmax_output, softmax_output)

# Plot Softmax function
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(z[0], softmax_output[0], label='Class 1')
plt.plot(z[0], softmax_output[1], label='Class 2')
plt.title('Softmax Function')
plt.xlabel('Input value')
plt.ylabel('Probability')
plt.legend()

# Plot Softmax derivative
plt.subplot(1, 2, 2)
plt.imshow(S, cmap='viridis', extent=[-2, 2, -2, 2])
plt.title('Softmax Derivative')
plt.xlabel('Input value class 1')
plt.ylabel('Input value class 2')
plt.colorbar()

plt.tight_layout()
plt.show()
```

```python
#softmax with derivatives
import numpy as np
import matplotlib.pyplot as plt

# Generate x values for plotting (assuming two classes for simplicity)
x = np.linspace(-5, 5, 100)
x_arr = np.stack((x, -x), axis=-1)  # create a 2D array for two-class example

# Compute softmax values
e_x = np.exp(x_arr - np.max(x_arr, axis=-1, keepdims=True))
softmax_y = e_x / np.sum(e_x, axis=-1, keepdims=True)

# Compute derivative values
softmax_derivative_y = softmax_y * (1 - softmax_y)

# Plot softmax function
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, softmax_y[:, 0], label='Class 1')
plt.plot(x, softmax_y[:, 1], label='Class 2')
plt.title('Softmax Activation Function')
plt.xlabel('x')
plt.ylabel('Probability')
plt.grid(True)
plt.legend()

# Plot derivative of softmax function
plt.subplot(1, 2, 2)
plt.plot(x, softmax_derivative_y[:, 0], label="Softmax'(Class 1)")
plt.plot(x, softmax_derivative_y[:, 1], label="Softmax'(Class 2)")
plt.title("Derivative of Softmax Activation Function")
plt.xlabel('x')
plt.ylabel("Derivative")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


```
